```python
df_test.isnull().sum()
```

```python
# Dropping the non-required columns

df_train.drop(['Unnamed: 0', 'merchant', 'trans_date_trans_time', 'street', 'city', 'first', 'last', 'job', 'trans_num'], axis = 1, inplace = True)
df_train
```

```python
df_test.drop(['Unnamed: 0', 'merchant', 'trans_date_trans_time', 'street', 'city', 'first', 'last', 'job', 'trans_num'], axis = 1, inplace = True)
df_test
```

```python
# Replacing the date of birth column with the age

from datetime import datetime
from dateutil.relativedelta import relativedelta

age = []

for x in range(0, df_train['dob'].size): 
    dob_date = datetime.strptime(df_train['dob'][x], "%Y-%m-%d")
    age_year = relativedelta(datetime.now(), dob_date)
    age.append(age_year.years)
```

```python
age
```

```python
df_train['age'] = age
df_train.drop('dob', axis = 1, inplace = True)
df_train
```

```python
age = []

for x in range(0, df_test['dob'].size): 
    dob_date = datetime.strptime(df_test['dob'][x], "%Y-%m-%d")
    age_year = relativedelta(datetime.now(), dob_date)
    age.append(age_year.years)
```

```python
age
```

```python
df_test['age'] = age
df_test.drop('dob', axis = 1, inplace = True)
df_test
```

```python
# Encoding the categorical variables

df_train.gender.replace(['F', 'M'], [1, 0], inplace = True)

df_test.gender.replace(['F', 'M'], [1, 0], inplace = True)
```

```python
dummies = pd.get_dummies(df_train[['category', 'state']], columns = ['category', 'state'], drop_first = True)
dummies = dummies.astype(int)
df_train.drop(['category', 'state'], axis = 'columns', inplace = True)
df_train = pd.concat([df_train, dummies], axis = 1)
df_train
```

```python
dummies = pd.get_dummies(df_test[['category', 'state']], columns = ['category', 'state'], drop_first = True)
dummies = dummies.astype(int)
df_test.drop(['category', 'state'], axis = 'columns', inplace = True)
df_test = pd.concat([df_test, dummies], axis = 1)
df_test
```

```python
# Scaling the data using the MinMaxScaler

from sklearn.preprocessing import MinMaxScaler

columns_to_scale = ['cc_num', 'amt', 'zip', 'lat', 'long', 'city_pop', 'unix_time', 'merch_lat', 'merch_long', 'age']

scaler = MinMaxScaler()
df_train[columns_to_scale] = scaler.fit_transform(df_train[columns_to_scale])

df_test[columns_to_scale] = scaler.fit_transform(df_test[columns_to_scale])
```

```python
df_train.columns
```

```python
# Splitting the data into dependant and independant features

X_train = df_train.drop('is_fraud', axis = 1)
y_train = df_train['is_fraud']

from imblearn.over_sampling import SMOTE

smote = SMOTE(sampling_strategy = 'minority')
X_smote, y_smote = smote.fit_resample(X_train, y_train)

y_smote.value_counts()
```

```python
X_test = df_test.drop('is_fraud', axis = 1)
y_test = df_test['is_fraud']
```

```python
# Defining classifiers and their parameters to use in the GridSearchCV

from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

random_forest = RandomForestClassifier()
random_forest_params = {
    'n_estimators': [50, 100],
    'criterion': ['gini', 'entropy']
}
```

```python
logistic_regression = LogisticRegression()
logistic_regression_params = {
    'penalty': ['elasticnet', 'none'],
    'C': [0.1, 1.0, 10.0]
}
```

```python
decision_tree = DecisionTreeClassifier()
decision_tree_params = {
    'max_features': ['log2', None],
    'criterion': ['gini', 'entropy'],
}
```

```python
classifiers = [
    (random_forest, random_forest_params),
    (logistic_regression, logistic_regression_params),
    (decision_tree, decision_tree_params)
]
```

```python
# Using the GridSearchCV to analyze the result with the different classifiers and their parameters

from sklearn.model_selection import GridSearchCV

scores = []

for classifier, params in classifiers:
    grid_search = GridSearchCV(classifier, params, cv = 5)
    grid_search.fit(X_smote, y_smote)
    
    scores.append({
        'model' : classifier,
        'best_score' : grid_search.best_score_,
        'best_params' : grid_search.best_params_
    })
```

```python
# From the result of the GridSearchCV getting the classifier with the best accuracy with its parameters

df = pd.DataFrame(scores, columns = ['model', 'best_score', 'best_params'])
df
```

```python
# Training the model with the best classifier that we got from the GridSearchCV 

decision_tree = DecisionTreeClassifier(criterion = 'entropy', max_features = 'log2')
model = decision_tree.fit(X_smote, y_smote)
```

```python
# Getting the accuracy of the model from the testing data

model.score(X_test, y_test)
```

```python
# To further analyze our model, storing the predicted data of the testing sample into a variable

y_predicted = model.predict(X_test)
y_predicted
```

```python
# Displaying a confustion matrix

import seaborn as sn
from sklearn.metrics import confusion_matrix
from sklearn import metrics

confusion_matrix = metrics.confusion_matrix(y_test, y_predicted)

plt.figure(figsize = (10, 7))
sn.heatmap(confusion_matrix, annot = True, fmt = 'd')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
